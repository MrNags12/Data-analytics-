```python
import pandas as pd
```

```python
df=pd.read_csv(r"/content/N.csv", on_bad_lines='skip')
```

```python
df['State'] = df['State'].str.strip().str.title()
```

```python
df['Crop'] = df['Crop'].str.strip().str.title()
```

```python
print(df.columns)
```

```python
mask_prod = df['Production'].isna() & df['Area '].notna() & df['Yield'].notna()
df.loc[mask_prod, 'Production'] = df.loc[mask_prod, 'Area '] * df.loc[mask_prod, 'Yield']
```

```python
df = df[df['Area '] > 0]
```

```python
df = df[df['Production'] >= 0]
```

```python
df = df.drop_duplicates()
```

```python
season_map = {'Autumn':'Autumn','Kharif':'Kharif','Summer':'Summer','Rabi':'Rabi'}
```

```python
df['Season'] = df['Season'].map(season_map).fillna(df['Season'])
```

```python
df['Yield_calc'] = df['Production'] / df['Area ']
```

```python
(df.loc[~df['Yield'].isna(), ['Yield','Yield_calc']].assign(diff=lambda x: x['Yield']-x['Yield_calc'])).head()
```

```python
df.groupby('Crop')['Production'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('State')['Production'].sum().sort_values(ascending=False).head(10)
```

```python
yearly = df.groupby('Crop_Year')['Production'].sum().reset_index()
```

```python
df.groupby(['Season','Crop'])['Yield'].mean().unstack().head()
```

```python
df.groupby(['State','District '])['Production'].sum().sort_values(ascending=False).head(20)
```

```python
df = df.sort_values(['State','Crop','Crop_Year'])
```

```python
df['Yield_YoY'] = df.groupby(['State','Crop'])['Yield'].pct_change()
```

```python
df.to_csv(r"/NEW")
```

```python
df['Yield_YoY'] = df.groupby(['State', 'Crop'])['Yield'].pct_change()
```

```python
df['Yield_YoY'] = df['Yield_YoY'].fillna(0)
```

```python
df = df.dropna(subset=['Yield_YoY'])
```

```python
df.isna().sum()
```

```python
df.to_csv(r"/content/N")
```

```python
import matplotlib.pyplot as plt
```

```python
yearly_prod = df.groupby('Crop_Year')['Production'].sum()
plt.figure(figsize=(10,5))
yearly_prod.plot(kind='bar', color='skyblue')
plt.title('Total Crop Production by Year')
plt.ylabel('Production (Tonnes)')
plt.xlabel('Year')
plt.tight_layout()
plt.show()
```

```python
top_crops = df.groupby('Crop')['Production'].sum().sort_values(ascending=False).head(10)
top_crops.plot(kind='bar', color='orange', figsize=(10,5))
plt.title('Top 10 Crops by Total Production')
plt.ylabel('Total Production (Tonnes)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
season_yield = df.groupby('Season')['Yield'].mean().sort_values()
season_yield.plot(kind='barh', color='green')
plt.title('Average Yield by Season')
plt.xlabel('Yield (tonnes/hectare)')
plt.tight_layout()
plt.show()
```

```python
crop_name = 'Rice'
crop_trend = df[df['Crop'] == crop_name].groupby('Crop_Year')['Yield'].mean()
crop_trend.plot(marker='o', color='purple', figsize=(10,5))
plt.title(f'Yield Trend Over Years: {crop_name}')
plt.ylabel('Avg Yield (tonnes/hectare)')
plt.xlabel('Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
latest_year = df['Crop_Year'].max()
state_prod = df[df['Crop_Year'] == latest_year].groupby('State')['Production'].sum().sort_values()
state_prod.plot(kind='barh', figsize=(10,8), color='steelblue')
plt.title(f'State-wise Total Production in {latest_year}')
plt.xlabel('Production (Tonnes)')
plt.tight_layout()
plt.show()
```

```python
yoy_mean = df.groupby('Crop')['Yield_YoY'].mean().dropna().sort_values(ascending=False).head(10)
yoy_mean.plot(kind='bar', color='crimson', figsize=(10,5))
plt.title('Top 10 Crops by Avg YoY Yield Growth')
plt.ylabel('Average YoY % Change')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['Crop_Year'] = pd.to_numeric(df['Crop_Year'], errors='coerce')
```

```python
import pandas as pd

# Use the existing DataFrame df

# Step 1: Find the most common crop for each season (ignoring NaNs in Crop)
season_major_crop = (
    df.dropna(subset=['Crop'])
    .groupby('Season')['Crop']
    .agg(lambda x: x.value_counts().idxmax())
    .to_dict()
)

print("Most common crop by season:", season_major_crop)

# Step 2: Create a new column and fill missing Crop values
df['Crop_filled'] = df['Crop']
df.loc[df['Crop_filled'].isna(), 'Crop_filled'] = df.loc[df['Crop_filled'].isna(), 'Season'].map(season_major_crop)

# Step 3: Save the updated dataset (optional)
df.to_csv("APY_filled.csv", index=False)

# Step 4: Check rows that were filled
filled_rows = df[df['Crop'].isna()]
print(filled_rows[['State', 'District ', 'Season', 'Crop', 'Crop_filled']])
```
